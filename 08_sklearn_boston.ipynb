```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston

# boston 객체 생성
boston = load_boston()
print(boston.DESCR)
```

```python
boston_df = pd.DataFrame(boston.data, columns=boston.feature_names)
boston_df.head()
```

```python
boston_df['PRICE'] = boston.target
boston_df.head()
```

```python
print('boston house prices dataset size =', boston_df.shape)
```

```python
boston_df.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# X, Y 분할하기
X = boston_df.drop(['PRICE'], axis=1, inplace=False)  # 독립 변수
Y = boston_df['PRICE']  # 종속 변수
```

```python
# 훈련용 데이터와 평가용 데이터 분할하기(훈련 7 : 평가 3)
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=156)
```

```python
# 선형 회귀 분석 : 모델 생성
lr = LinearRegression()

# 선형 회귀 분석 : 모델 훈련
lr.fit(X_train, Y_train)

# 선형 회귀 분석 : 평가 데이터에 대한 예측 수행 → 예측 결과 Y_predict 구하기
Y_predict = lr.predict(X_test)
```

```python
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)

print('MSE = {0:.3f}, RMSE = {1:.3f}'.format(mse, rmse))
print('R^2(Variance score) = {0:.3f}'.format(r2_score(Y_test, Y_predict)))
print('Y 절편 값 =', lr.intercept_)
print('회귀 계수 값 =', np.round(lr.coef_, 1))
```

```python
coef = pd.Series(data=np.round(lr.coef_, 2), index=X.columns)
coef.sort_values(ascending=False)
```
