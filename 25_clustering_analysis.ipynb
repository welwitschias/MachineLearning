```python
import pandas as pd

retail_df = pd.read_excel('data/Online_Retail.xlsx')
retail_df.head()
```

```python
# 정보 확인
retail_df.info()
```

```python
# 데이터 정제하기
retail_df = retail_df[retail_df['Quantity'] > 0]  # 음수 제거
retail_df = retail_df[retail_df['UnitPrice'] > 0]  # 음수 제거
retail_df = retail_df[retail_df['CustomerID'].notnull()]  # null 제거
retail_df['CustomerID'] = retail_df['CustomerID'].astype(int)  # 정수형으로 변환
```

```python
retail_df.info()
```

```python
retail_df.isnull().sum()
```

```python
retail_df.shape
```

```python
# 중복 레코드 제거
retail_df.drop_duplicates(inplace=True)
```

```python
retail_df.shape
```

```python
# column 추출 및 분석용 데이터 생성
pd.DataFrame([{'Product': len(retail_df['StockCode'].value_counts()),  # 제품수
               'Transaction': len(retail_df['InvoiceNo'].value_counts()),  # 거래건수
               'Customer': len(retail_df['CustomerID'].value_counts())}],  # 고객수
             columns=['Product', 'Transaction', 'Customer'], index=['counts'])
```

```python
retail_df['Country'].value_counts()
```

```python
# 주문 금액 column 추가
retail_df['SaleAmount'] = retail_df['UnitPrice'] * retail_df['Quantity']
retail_df.head()
```

```python
aggregations = {'InvoiceNo': 'count',  # 주문건수(갯수 count)
                'SaleAmount': 'sum',  # 주문금액(총액 sum)
                'InvoiceDate': 'max'  # 주문일(최근날짜 max)
                }
customer_df = retail_df.groupby('CustomerID').agg(aggregations)
customer_df = customer_df.reset_index()
customer_df.head()
```

```python
# column명 변경
customer_df = customer_df.rename(  
    columns={'InvoiceNo': 'Frequency', 'InvoiceDate': 'ElapsedDays'})
customer_df.head()
```

```python
import datetime

# 기준일 - 마지막 구매일 계산(기준일 : 2011-12-10)
customer_df['ElapsedDays'] = datetime.datetime(
    2011, 12, 10) - customer_df['ElapsedDays']
customer_df.head()
```

```python
# 마지막 구매 후 며칠이 지났는지를 일수로 환산
customer_df['ElapsedDays'] = customer_df['ElapsedDays'].apply(
    lambda x: x.days+1)
customer_df.head()
```

```python
import matplotlib.pyplot as plt

# boxplot 그리기
fig, ax = plt.subplots()
ax.boxplot([customer_df['Frequency'], customer_df['SaleAmount'],
            customer_df['ElapsedDays']], sym='bo')
plt.xticks([1, 2, 3], ['Frequency', 'SaleAmount', 'ElapsedDays'])
plt.show()

# SaleAmount 값이 너무 크므로 scale 조정이 필요함
```

```python
import numpy as np

# 로그함수 적용
customer_df['Freq_log'] = np.log1p(customer_df['Frequency'])
customer_df['SaleAmount_log'] = np.log1p(customer_df['SaleAmount'])
customer_df['ElapsedDays_log'] = np.log1p(customer_df['ElapsedDays'])
customer_df.head()
```

```python
fig, ax = plt.subplots()
ax.boxplot([customer_df['Freq_log'], customer_df['SaleAmount_log'],
            customer_df['ElapsedDays_log']], sym='bo')
plt.xticks([1, 2, 3], ['Freq_log', 'SaleAmount_log', 'ElapsedDays_log'])
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# X_features를 정규분포로 스케일링하기
X_features = customer_df[['Freq_log', 'SaleAmount_log', 'ElapsedDays_log']].values
X_features_scaled = StandardScaler().fit_transform(X_features)
```

```python
# 엘보 방법으로 클러스터 개수 k 선택하기
distortions = []
for i in range(1, 11):
    kmeans_i = KMeans(n_clusters=i, random_state=0)  # k평균 모델 생성
    kmeans_i.fit(X_features_scaled)  # k평균 모델 훈련
    distortions.append(kmeans_i.inertia_)  # inertia_ : 왜곡 값

plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of Cluster')
plt.ylabel('Distortion')
plt.show()
```

```python
# 클러스터 개수 k=3 설정 후 k평균 모델 재구축
kmeans = KMeans(n_clusters=3, random_state=0)  # k평균 모델 재생성

# 모델 학습 결과 예측(클러스터 레이블 생성)
Y_labels = kmeans.fit_predict(X_features_scaled)
customer_df['ClusterLabel'] = Y_labels
customer_df.head()
```

```python
# 각 클러스터의 비중을 bar chart로 시각화
from matplotlib import cm
from sklearn.metrics import silhouette_score, silhouette_samples


def silhouetteViz(n_cluster, X_features):
    kmeans = KMeans(n_clusters=n_cluster, random_state=0)
    Y_labels = kmeans.fit_predict(X_features)
    silhouette_values = silhouette_samples(
        X_features, Y_labels, metric='euclidean')

    y_ax_lower, y_ax_upper = 0, 0
    y_ticks = []

    for c in range(n_cluster):
        c_silhouettes = silhouette_values[Y_labels == c]
        c_silhouettes.sort()
        y_ax_upper += len(c_silhouettes)
        color = cm.jet(float(c)/n_cluster)
        plt.barh(range(y_ax_lower, y_ax_upper), c_silhouettes,
                 height=1.0, edgecolor='none', color=color)
        y_ticks.append((y_ax_lower + y_ax_upper)/2.0)
        y_ax_lower += len(c_silhouettes)
        silhouette_avg = np.mean(silhouette_values)

    plt.axvline(silhouette_avg, color='red', linestyle='--')
    plt.title('Number of Cluster : ' + str(n_cluster) + '\n' +
              'Silhouette Score : ' + str(round(silhouette_avg, 3)))
    plt.xticks([0, 0.2, 0.4, 0.6, 0.8, 1])
    plt.yticks(y_ticks, range(n_cluster))
    plt.xlabel('Silhouette Coefficient')
    plt.ylabel('Cluster')
    plt.tight_layout()
    plt.show()
```

```python
silhouetteViz(3, X_features_scaled)
```

```python
silhouetteViz(4, X_features_scaled)
```

```python
silhouetteViz(5, X_features_scaled)
```

```python
silhouetteViz(6, X_features_scaled)
```

```python
# 클러스터 데이터 분포 확인을 위한 scatter chart 시각화
def clusterScatter(n_cluster, X_features):
    c_colors = []
    kmeans = KMeans(n_clusters=n_cluster, random_state=0)
    Y_labels = kmeans.fit_predict(X_features)

    for i in range(n_cluster):
        c_color = cm.jet(float(i)/n_cluster)
        c_colors.append(c_color)

        # 클러스터의 데이터 분포를 동그라미로 시각화
        plt.scatter(X_features[Y_labels == i, 0], X_features[Y_labels == i, 1],
                    marker='o', color=c_color, edgecolors='black', s=50, label='cluster'+str(i))

    # 각 클러스터의 중심점을 삼각형으로 표시
    for i in range(n_cluster):
        plt.scatter(kmeans.cluster_centers_[i, 0], kmeans.cluster_centers_[i, 1],
                    marker='^', color=c_colors[i], edgecolors='w', s=200)
        
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()
```

```python
clusterScatter(3, X_features_scaled)
```

```python
clusterScatter(4, X_features_scaled)
```

```python
clusterScatter(5, X_features_scaled)
```
